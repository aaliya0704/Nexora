```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
messages = [
    {
        "role": "system",
        "content": "You are Nexora, a helpful, friendly, and intelligent AI assistant."
    },
    {
        "role": "user",
        "content": "Hello"
    }
]

prompt = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)

print(prompt)
```

```python
inputs = tokenizer(prompt, return_tensors="pt")

output = model.generate(
    **inputs,
    max_new_tokens=100,
    do_sample=True,
    temperature=0.7
)

response = tokenizer.decode(
    output[0],
    skip_special_tokens=True
)

print(response)
```

```python
def chat_response(user_input):
    # Step 1: Create the conversation
    messages = [
        {
            "role": "system",
            "content": "You are Nexora, a helpful, friendly, and intelligent AI assistant."
        },
        {
            "role": "user",
            "content": user_input
        }
    ]

    # Step 2: Convert the conversation into the model's chat format
    prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    # Step 3: Convert text into tokens
    inputs = tokenizer(prompt, return_tensors="pt")

    # Step 4: Generate a response
    output = model.generate(
        **inputs,
        max_new_tokens=100,
        do_sample=True,
        temperature=0.7
    )

    # Step 5: Keep only the newly generated tokens
    generated_tokens = output[0][inputs["input_ids"].shape[1]:]

    # Step 6: Convert tokens back into readable text
    response = tokenizer.decode(
        generated_tokens,
        skip_special_tokens=True
    )

    return response.strip()
```

```python
print(chat_response("Hello"))
```
