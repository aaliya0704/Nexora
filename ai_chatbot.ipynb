```python
def chat_response(user_input):
    result = chatbot(
        user_input,
        max_new_tokens=50,
        pad_token_id=50256
    )

    response = result[0]["generated_text"]

    return response.strip()
```

```python
print(chat_response("What is Artificial Intelligence?"))
```
